## Dataset Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

churn_data = pd.read_csv('churn_data.csv')
customer_data = pd.read_csv('customer_data.csv')
internet_data = pd.read_csv('internet_data.csv')

merged_data = pd.merge(churn_data, customer_data, on='customerID', how='inner')
final_merged_data = pd.merge(merged_data, internet_data, on='customerID', how='inner')

final_merged_data.to_csv('merged_data.csv', index=False)
```

```python
final_merged_data
```

```python
df = final_merged_data
```

```python
df.replace(" ", np.nan, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df
```

## Plotting Functions

```python
def lineplot(colname):
    value_counts = df[colname].value_counts().sort_index()
    plt.plot(value_counts)
    plt.title('Value Counts of '+colname)
    plt.xlabel(colname)
    plt.ylabel('Counts')
    plt.grid(True)
    plt.show()

def areaplot(colname):
    value_counts = df[colname].value_counts().sort_index()
    value_counts.plot(kind='area')
    plt.title('Value Counts of '+colname)
    plt.xlabel(colname)
    plt.ylabel('Counts')
    plt.grid(True)
    plt.show()

def barplot(colname):
    value_counts = df[colname].value_counts()
    value_counts.plot(kind='bar')
    plt.title('Counts of '+colname)
    plt.xlabel(colname)
    plt.ylabel('Counts')
    plt.grid(True)
    plt.show()
```

## Data Levelling

```python
def ConvertToNum(colname):
    codes, uniques = pd.factorize(df[colname])
    return codes
```

```python
df['PhoneService'] = ConvertToNum('PhoneService')
df['Contract'] = ConvertToNum('Contract')
df['PaperlessBilling'] = ConvertToNum('PaperlessBilling')
df['gender'] = ConvertToNum('gender')
df['Partner'] = ConvertToNum('Partner')
df['Dependents'] = ConvertToNum('Dependents')
df['MultipleLines'] = ConvertToNum('MultipleLines')
df['InternetService'] = ConvertToNum('InternetService')
df['OnlineSecurity'] = ConvertToNum('OnlineSecurity')
df['OnlineBackup'] = ConvertToNum('OnlineBackup')
df['DeviceProtection'] = ConvertToNum('DeviceProtection')
df['TechSupport'] = ConvertToNum('TechSupport')
df['StreamingTV'] = ConvertToNum('StreamingTV')
df['StreamingMovies'] = ConvertToNum('StreamingMovies')
df['PaymentMethod'] = ConvertToNum('PaymentMethod')
df['Churn'] = ConvertToNum('Churn')
```

## Data training

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint as sp_randint
import seaborn as sns
```

```python
def showcm(cm):
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title('Confusion Matrix')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()
```

```python
model = [] 
acc = []
model.append('Decision Tree')
model.append('Random Forest')
model.append('Logistic Regression')
model.append('SVM')
model
```

```python
prec = []
rec = []
f = []
```

```python
X = df.drop(['customerID', 'Churn'], axis=1)
y = df['Churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

### Decision Tree Classifier

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
cm = confusion_matrix(y_test, y_pred)
showcm(cm)
```

```python
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
```

```python
prec.append(precision)
rec.append(recall)
f.append(f1)
```

```python
accuracy
```

```python
acc.append(accuracy)
```

### Random Forest Classifier

```python
clf1 = RandomForestClassifier(random_state=42)
```

```python
param_dist = {
    "n_estimators": sp_randint(100, 500),
    "max_depth": [3, None],
    "max_features": sp_randint(1, 11),
    "min_samples_split": sp_randint(2, 11),
    "min_samples_leaf": sp_randint(1, 11),
    "bootstrap": [True, False]
}
```

```python
n_iter_search = 20  # Number of iterations
random_search = RandomizedSearchCV(clf1, param_distributions=param_dist, cv=5, random_state=42)
```

```python
random_search.fit(X_train, y_train)

print("Best parameters found: ", random_search.best_params_)
print("Best score: ", random_search.best_score_)
```

```python
fm = RandomForestClassifier(n_estimators=419, max_depth=None, max_features=3, min_samples_leaf=7, min_samples_split=5, bootstrap=True, max_samples=0.6)
```

```python
fm.fit(X_train, y_train)
```

```python
y_pred1 = fm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred1)
accuracy
```

```python
precision = precision_score(y_test, y_pred1)
recall = recall_score(y_test, y_pred1)
f1 = f1_score(y_test, y_pred1)
prec.append(precision)
rec.append(recall)
f.append(f1)
```

```python
cm = confusion_matrix(y_test, y_pred1)
showcm(cm)
```

```python
acc.append(accuracy)
```
